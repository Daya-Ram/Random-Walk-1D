### 1D Random Walk

```python
import numpy as np
from itertools import cycle
import matplotlib.pyplot as plt
import random as random
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline


# Define parameters for the walk
dims = 1
step_n = 100800
step_set = [-1, 0, 1]
origin = np.zeros((1,dims))

# Simulate steps in 1D
step_shape = (step_n,dims)
steps = np.random.choice(a=step_set, size=step_shape)
path = np.concatenate([origin, steps]).cumsum(0)
start = path[:1]
stop = path[-1:]

# Plot the path
fig = plt.figure(figsize=(18,14),dpi=200)
ax = fig.add_subplot(111)
ax.scatter(np.arange(step_n+1), path, c='maroon', alpha=0.25,s=0.05);
ax.plot(path,c='blue', alpha=0.5,lw=0.5,ls= '-',);
ax.plot(0, start, c='red', marker= '+')
ax.plot(step_n, stop, c='black', marker='o')
plt.title('One Domensional Random Walk')
plt.tight_layout(pad=0)
plt.show()
```

### Random Walk in 1D

import random as random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

```python
#probability to move up or down
prob= [0.2,0.4]
# statically defining the starting position
start=0
positions=[start]
#creating the random points
rr=np.random.random(100)
downp=rr<prob[0]
upp=rr>prob[1]

for idownp, iupp in zip(downp,upp):
    down=idownp and positions [-1]>1
    up=iupp and positions [-1]<2
    positions.append(positions[-1]-2*down+3*up)
    # plotting down the graph of the random walk in 1D
    plt.plot(positions)
plt.show()
```
